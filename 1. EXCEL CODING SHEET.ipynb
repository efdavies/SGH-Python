```python
import tkinter as tk
from tkinter import PhotoImage
import os
from openpyxl import Workbook
from openpyxl.drawing.image import Image as OpenpyxlImage
from openpyxl.styles import Font, PatternFill, Border, Side
from win32com.propsys import propsys
from win32com.shell import shellcon

def SGH_template(ws):
    bold_font = Font(bold=True)
    white_fill = PatternFill(start_color='FFFFFFFF', end_color='FFFFFFFF', fill_type='solid')
    grey_fill = PatternFill(start_color="808080", end_color="808080", fill_type="solid")
    right_border = Border(right=Side(style='thin', color='000000'))
    left_border = Border(left=Side(style='thin', color='000000'))
    top_border = Border(top=Side(style='thin', color='000000'))
    bottom_border = Border(bottom=Side(style='thin', color='000000'))

    for row in ws['A1:J6']:
        for cell in row:
            cell.fill = white_fill
    
    for row in ws['A9:J9']:
        for cell in row:
            cell.fill = white_fill

    columns = ['K', 'L', 'M', 'N', 'O', 'P', 'Q', 'R', 'S', 'T', 'U', 'V', 'W', 'X', 'Y', 'Z', 'AA','AB', 'AC', 'AD','AE', 'AF', 'AG', 'AH', 'AI', 'AJ', 'AK', 'AL', 'AM']
    for col in columns:
        for row in range(1, 1001):
            cell = ws[col + str(row)]
            cell.fill = grey_fill

    ws['A1'].border = Border(top=Side(style='thin', color='000000'), left=Side(style='thin', color='000000'))
    ws['A6'].border = Border(bottom=Side(style='thin', color='000000'), left=Side(style='thin', color='000000'))
    ws['J1'].border = Border(top=Side(style='thin', color='000000'), right=Side(style='thin', color='000000'))
    ws['J6'].border = Border(bottom=Side(style='thin', color='000000'), right=Side(style='thin', color='000000'))

    for row in ws['B1:I1']:
        for cell in row:
            cell.border = top_border

    for row in ws['B6:I6']:
        for cell in row:
            cell.border = bottom_border

    for col in ws['A2:A5']:
        for cell in col:
            cell.border = left_border

    for col in ws['J2:J5']:
        for cell in col:
            cell.border = Border(bottom=Side(style='thin', color='000000'), right=Side(style='thin', color='000000'))

    for cell in ['B4', 'B5', 'C4', 'C5', 'D4', 'D5', 'E4', 'E5', 'F4', 'F5', 'G4', 'G5', 'H4', 'H5']:
        ws[cell].border = bottom_border

    img = OpenpyxlImage('SGH Logo.png')
    img.anchor = 'A1'
    ws.add_image(img)

    ws["A4"] = "Client"
    ws["A4"].font = bold_font

    ws["A5"] = "Subject"
    ws["A5"].font = bold_font

    ws["I2"] = "Sheet NO."
    ws["I2"].font = bold_font

    ws["I3"] = "Project NO"
    ws["I3"].font = bold_font

    ws["I4"] = "Date"
    ws["I4"].font = bold_font

    ws["I5"] = "By"
    ws["I5"].font = bold_font

    ws["I6"] = "Client"
    ws["I6"].font = bold_font

def format_worksheet(ws, path):
    bold_font = Font(bold=True)
    ws["A7"] = "File Path"
    ws["A8"] = f"{path}"
    ws["A7"].font = bold_font

    ws.merge_cells("A9:C9")
    ws["A10"] = "Photo"
    ws["A10"].font = bold_font

    ws["C10"] = "Tag 1"
    ws["C10"].font = bold_font

    ws["D10"] = "Tag 2"
    ws["D10"].font = bold_font

    ws["E10"] = "Tag 3"
    ws["E10"].font = bold_font

    ws["F10"] = "Tag 4"
    ws["F10"].font = bold_font

    ws.merge_cells("G10:H10")
    ws["G10"] = "New Folders"
    ws["G10"].font = bold_font

    ws["A8"] = path

def extract_keywords(image_path):
    pk = propsys.PSGetPropertyKeyFromName("System.Keywords")
    ps = propsys.SHGetPropertyStoreFromParsingName(image_path, None, shellcon.GPS_READWRITE, propsys.IID_IPropertyStore)
    keywords = ps.GetValue(pk).GetValue()
    return keywords

def create_workbook(file_path, file_name):
    path = file_path  
    wb = Workbook()
    ws = wb.active
    SGH_template(ws)
    ws.title = 'Image Keywords'
    
    format_worksheet(ws, path)

    if os.path.exists(path):
        row = 11
        for file in os.listdir(path):
            if file.lower().endswith(".jpg"):
                img_path = os.path.join(path, file)
                ws.cell(column=1, row=row, value=f'=HYPERLINK("{img_path}", "{file}")')
                
                keywords = extract_keywords(img_path)
                
                if keywords:
                    ws.cell(row=row, column=3, value=keywords[0])
                    
                    if len(keywords) > 1:
                        ws.cell(row=row, column=4, value=keywords[1])
                    
                    if len(keywords) > 2:
                        ws.cell(row=row, column=5, value=keywords[2]) 

                    if len(keywords) > 3:
                        ws.cell(row=row, column=6, value=keywords[3])                              
                row += 1
    else:
        print(f"The file path {path} does not exist. Please check the file path.")
    
    try:
        # Ensure the file is saved regardless of whether it existed before
        file_path = f"{file_name}.xlsx"
        wb.save(file_path)
        # Return the absolute path of the saved file
        return os.path.abspath(file_path)
    except PermissionError:
        print(f"Permission denied: unable to save '{file_name}.xlsx'. Please close the file if it's open, or try saving to a different location.")
        return None

def on_enter():
    file_name = entry_file_name.get().strip()
    file_path = entry_file_path.get().strip()
    if file_name and file_path:
        excel_file_path = create_workbook(file_path, file_name)
        if excel_file_path:
            result_label.config(text=f"Workbook {file_name}.xlsx created successfully at:\n{excel_file_path}")
            # Open the Excel file
            os.startfile(excel_file_path)
            # Open the folder containing the Excel file
            os.startfile(os.path.dirname(excel_file_path))
        else:
            result_label.config(text="Failed to create the workbook.")
    else:
        result_label.config(text="Please enter both the file name and file path.")

def create_ui():
    global entry_file_name, entry_file_path, result_label

    # Create the main window
    root = tk.Tk()
    root.title("Workbook Creator")
    root.geometry("600x400")  # Set the window size to 600x400 pixels

    # Load the logo image
    logo_image = PhotoImage(file="SGH Logo.png")

    # Create and place the widgets
    logo_label = tk.Label(root, image=logo_image)
    logo_label.pack(pady=0)

    label_file_name = tk.Label(root, text="Enter Spreadsheet Name:")
    label_file_name.pack(pady=10)

    entry_file_name = tk.Entry(root, width=50)
    entry_file_name.pack(pady=10)

    label_file_path = tk.Label(root, text="Enter File Path:")
    label_file_path.pack(pady=10)

    entry_file_path = tk.Entry(root, width=50)
    entry_file_path.pack(pady=10)

    enter_button = tk.Button(root, text="Enter", command=on_enter)
    enter_button.pack(pady=10)

    result_label = tk.Label(root, text="", wraplength=500)
    result_label.pack(pady=10)

    # Run the GUI event loop
    root.mainloop()

# Call the function to create the UI
create_ui()
```

The Base directory is the filepath that you want your new folder to be placed in

```python
import tkinter as tk
from tkinter import PhotoImage
import os
import shutil
from openpyxl import load_workbook
from urllib.parse import quote

def organization(xl_file, base_dir):
    # Define folder and path
    parent_folder_name = "Structures"
    parent_dir = os.path.join(base_dir, parent_folder_name)

    # Load the workbook and active worksheet
    wb = load_workbook(f'{xl_file}.xlsx')
    ws = wb.active

    # Ensure the parent directory exists
    os.makedirs(parent_dir, exist_ok=True)

    # Update base_dir to the new parent directory
    base_dir = parent_dir

    # Keep track of created folders to avoid duplicates
    created_folders = set()

    # Process the spreadsheet
    for i in range(11, ws.max_row + 1):
        image_path = ws.cell(row=i, column=1).value  # Column A
        categories = [ws.cell(row=i, column=col).value for col in range(3, 7)]  # Columns C, D, E, F

        if image_path is None:
            continue

        if image_path.startswith('=HYPERLINK'):
            image_path = image_path.split('\"')[1]

        for category in categories:
            if category:
                new_folder_path = os.path.join(base_dir, category)
                if new_folder_path not in created_folders:
                    os.makedirs(new_folder_path, exist_ok=True)
                    created_folders.add(new_folder_path)

                new_image_path = os.path.join(new_folder_path, os.path.basename(image_path))
                if not os.path.exists(new_image_path):
                    shutil.copy(image_path, new_image_path)

    # Update the "New folders" column with the new folder paths
    existing_paths = []
    row = 11
    for folder_path in created_folders:
        if folder_path not in existing_paths:
            ws.cell(row=row, column=7, value=f'=HYPERLINK("{folder_path}", "{folder_path}")')  # Column G
            existing_paths.append(folder_path)
            row += 1

    # Save the changes to the workbook with error handling
    try:
        wb.save(f'{xl_file}.xlsx')
        print("Images have been moved to new folders and the spreadsheet has been updated.")
        return base_dir  # Return the path of the "Structures" folder
    except PermissionError:
        return f"Permission denied: unable to save '{xl_file}.xlsx'. Please close the file if it's open, and then re-run the code."

def on_enter():
    xl_file = entry_xl_file.get().strip()
    base_dir = entry_base_dir.get().strip()
    if xl_file and base_dir:
        result = organization(xl_file, base_dir)
        result_label.config(text=f"New 'Structures' folder path:\n{result}")
    else:
        result_label.config(text="Please enter both the spreadsheet name and base directory.")

def create_ui():
    global entry_xl_file, entry_base_dir, result_label

    # Create the main window
    root = tk.Tk()
    root.title("Image Organization Tool")
    root.geometry("600x400")  # Set the window size to 600x400 pixels

    # Load the logo image
    logo_image = PhotoImage(file="SGH Logo.png")

    # Create and place the widgets
    logo_label = tk.Label(root, image=logo_image)
    logo_label.pack(pady=10)

    label_xl_file = tk.Label(root, text="Enter Spreadsheet Name:")
    label_xl_file.pack(pady=10)

    entry_xl_file = tk.Entry(root, width=50)
    entry_xl_file.pack(pady=10)

    label_base_dir = tk.Label(root, text="Enter Base Directory:")
    label_base_dir.pack(pady=10)

    entry_base_dir = tk.Entry(root, width=50)
    entry_base_dir.pack(pady=10)

    enter_button = tk.Button(root, text="Enter", command=on_enter)
    enter_button.pack(pady=10)

    result_label = tk.Label(root, text="", wraplength=500)
    result_label.pack(pady=10)

    # Run the GUI event loop
    root.mainloop()

# Call the function to create the UI
create_ui()
```
